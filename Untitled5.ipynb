```python
import pandas as pd

df = pd.read_csv("titanic.csv")


print(df.head(8))
df.info()
# Пропущенные значения обнаружены в столбцах:
# Age, Cabin, Embarked

print(df.describe())
# По результатам describe():
# Средний возраст пассажиров = 29.7 лет
# Максимальная цена за билет = 512.3292
```

```python
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)

df.info()

def get_age_group(age):
    if age < 18:
        return "Ребенок"
    elif age <= 65:
        return "Взрослый"
    else:
        return "Пожилой"

df["AgeGroup"] = df["Age"].apply(get_age_group)

print(df.head())
```

```python
print(df.groupby("Sex")["Survived"].mean() * 100)

print(df.groupby("Pclass")["Survived"].mean() * 100)

print(
    df.pivot_table(
        values="Survived",
        index="Sex",
        columns="Pclass",
        aggfunc="mean"
    ) * 100
)

print(
    df[
        (df["Age"] < 18) &
        (df["Pclass"] == 3) &
        (df["Survived"] == 1)
    ][["Name", "Age"]].sort_values(by="Age", ascending=False)
)

print(
    df[
        (df["Sex"] == "male") &
        (df["Survived"] == 0)
    ].sort_values(by="Age", ascending=False).head(1)[["Name", "Age"]]
)
```
